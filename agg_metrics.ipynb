```python
import pandas as pd
import matplotlib.pyplot as plt

# Load CSV (adjust path as per your environment)
df = pd.read_csv("aggregated_metrics.csv")

# Correct any header issues
df.columns = ['metric', 'date', 'average_value', 'min_value', 'max_value']

# Convert 'date' to datetime
df['date'] = pd.to_datetime(df['date'], format='%Y-%m-%d')


# Display the DataFrame
print(df)

# Separate temperature and precipitation for plotting
temp_df = df[df['metric'].str.contains('temp', case=False)]
precip_df = df[df['metric'].str.contains('precip', case=False)]

# Plot Temperature
if not temp_df.empty:
    plt.figure(figsize=(6, 4))
    plt.plot(temp_df['date'], temp_df['average_value'], marker='o', color='orange', label='Avg Temp')
    plt.fill_between(temp_df['date'], temp_df['min_value'], temp_df['max_value'], color='orange', alpha=0.2, label='Min-Max Range')
    plt.title('Average Temperature Over Time')
    plt.xlabel('Date')
    plt.ylabel('Temperature (°F)')
    plt.legend()
    plt.grid(True)
    plt.show()

# Plot Precipitation
if not precip_df.empty:
    plt.figure(figsize=(6, 4))
    plt.plot(precip_df['date'], precip_df['average_value'], marker='o', color='blue', label='Avg Precipitation')
    plt.fill_between(precip_df['date'], precip_df['min_value'], precip_df['max_value'], color='blue', alpha=0.2, label='Min-Max Range')
    plt.title('Average Precipitation Over Time')
    plt.xlabel('Date')
    plt.ylabel('Precipitation (inches)')
    plt.legend()
    plt.grid(True)
    plt.show()
```
